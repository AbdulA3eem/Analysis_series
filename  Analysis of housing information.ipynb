### [Problem 5] Researching terminology

#### Skewness: 
+ is a measure of symmetry, or more precisely, the lack of symmetry. A distribution, or data set, is symmetric if it looks the same to the left and right of the center point.

#### Kurtosis: 
+ is a statistical measure that defines how heavily the tails of a distribution differ from the tails of a normal distribution. In other words, kurtosis identifies whether the tails of a given distribution contain extreme values.

### [Problem 6] Confirming distribution

```python
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis

# Replace inf values with NaN
train_data.replace([np.inf, -np.inf], np.nan, inplace=True)

# Calculate kurtosis and skewness for the original SalePrice
original_kurtosis = kurtosis(train_data['SalePrice'].dropna())
original_skewness = skew(train_data['SalePrice'].dropna())

# Plot the distribution of SalePrice
sns.set_theme()
np.random.seed(0)
plt.figure(figsize=(10, 6))
sns.histplot(train_data['SalePrice'], kde=True, bins=30)
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()

print(f'Original Kurtosis: {original_kurtosis}')
print(f'Original Skewness: {original_skewness}')
```

##### Log Transformation

```python
# Log transformation of SalePrice
train_data['Log_SalePrice'] = np.log(train_data['SalePrice'])

# Calculating the kurtosis and skewness for the transformed SalePrice
log_kurtosis = kurtosis(train_data['Log_SalePrice'].dropna())
log_skewness = skew(train_data['Log_SalePrice'].dropna())

# Ploting the distribution of Log_SalePrice
plt.figure(figsize=(10, 6))
sns.histplot(train_data['Log_SalePrice'], kde=True, bins=30)
plt.title('Distribution of Log-Transformed Sale Prices')
plt.xlabel('Log Sale Price')
plt.ylabel('Frequency')
plt.show()

print(f'Log-Transformed Kurtosis: {log_kurtosis}')
print(f'Log-Transformed Skewness: {log_skewness}')

# Comparison
print(f'Kurtosis before transformation: {original_kurtosis}')
print(f'Kurtosis after transformation: {log_kurtosis}')
print(f'Skewness before transformation: {original_skewness}')
print(f'Skewness after transformation: {log_skewness}')

# Explaining the change before and after logarithmic transformation

if log_skewness < original_skewness:
    print("Logarithmic transformation has reduced the skewness, making the distribution more symmetrical.")
else:
    print("Logarithmic transformation has not reduced the skewness.")

if log_kurtosis < original_kurtosis:
    print("Logarithmic transformation has reduced the kurtosis, indicating fewer extreme outliers.")
else:
    print("Logarithmic transformation has not reduced the kurtosis.")
```

### [Problem 7] Confirming the correlation coefficient

```python
# Creating a heat map of the correlation coefficient

numeric_df = train_data.select_dtypes(include=['number'])

# Computing the correlation 
corr_matrix = numeric_df.corr()

# Creating a heatmap of the correlation coefficient 
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Coefficient Matrix Heatmap')
plt.show()
```

```python
# 10 features that have a high correlation with the target variable

corr_target = corr_matrix[['SalePrice']].sort_values(by=['SalePrice'], ascending=False)
corr_target[2:12]
```

```python
# creating a correlation coefficient {matrix heat map}
target_corr_matrix = train_data[corr_target[2:12].index]
display(target_corr_matrix)
```

```python
plt.subplots(figsize=(15,12))
plt.title('Correlatiom matrix')
sns.heatmap(target_corr_matrix.corr(), vmax=1, annot=True, cmap="YlGnBu")
plt.show()
```
